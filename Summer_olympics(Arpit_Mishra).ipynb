# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv('summer.csv')
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
d={}
lst=[]
#for i in df['Sport'].unique():
df['Sport'].unique()
for i in df[df['Medal']=='Gold'].values: 
    if i[2] not in d:
        d[i[2]]=0
    d[i[2]]+=1
for i in d:
    lst.append((i,d[i]))
res=sorted(lst, key= lambda x:x[1],reverse=True)[:5]
res

```

```python
x=[]
y=[]
data=[]
for i in res:
    x.append(i[0])
    y.append(i[1])
plt.xlabel('Sport')
plt.ylabel('Gold freq')
plt.bar(x,y)
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df["Sport"].value_counts()[:5]
```

```python
df["Sport"].value_counts()[:5].plot(kind='bar')
plt.xlabel('Sports')
plt.ylabel('Medal freq')
plt.title('Plot')
```

### 4. Which player has won most number of medals? (Top 5)

```python
df['Athlete'].value_counts()[:5]
```

```python
df['Athlete'].value_counts()[:5].plot(kind='bar')
plt.xlabel('Athletes/Players')
plt.ylabel('Medal freq')
plt.title('Players with most medals')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)
